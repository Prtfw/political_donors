```python
import sys
from operator import itemgetter
from itertools import islice, groupby, tee, takewhile
from pprint import pprint
from statistics import median
from decimal import Decimal, ROUND_HALF_UP
from bisect import insort
from collections import defaultdict
import pdb
from timeit import default_timer as timer
```

```python

def rounder(x):
    return Decimal(x).quantize(0, ROUND_HALF_UP)
# tested
```

```python
def get_median(arr): #bench mark show this as slower, decided to use median from statistics instead
    it = map(float, arr)
    li = list(it)
    if len(li)%2 == 0:
        res = (li[len(li)//2]+li[len(li)//2-1])/2
    else:
        res = (li[len(li)//2])
    return int(rounder(res))

# tested but slow
```
